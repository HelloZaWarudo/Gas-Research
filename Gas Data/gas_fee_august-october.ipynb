```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.dates as mdates
```

```python
df = pd.read_csv('C:\Gas Research\gas_august-october.csv', parse_dates=['Date'])
```

```python
# Display the first few rows of the DataFrame
df.head()

# Get information about the DataFrame
df.info()

# Get basic statistical summaries of the DataFrame
df.describe()
```

```python
# Example: Plotting a histogram
df['Price'].hist()

plt.xlabel('Base Fee')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis
plt.title('Base Fee Histogram In Three Months')  # Title of the graph

plt.show()

# Example: Creating a scatter plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed

# Plotting
plt.plot(df['Date'], df['Price'])  # Ensure these are your actual column names

locator = mdates.DayLocator(interval=7)  # Adjust interval as needed
plt.gca().xaxis.set_major_locator(locator)
formatter = mdates.DateFormatter('%Y-%m-%d')
plt.gca().xaxis.set_major_formatter(formatter)
plt.xticks(rotation=45)

# Adding labels and title
plt.xlabel('Date')  # Label for x-axis
plt.ylabel('Price')  # Label for y-axis
plt.title('Price over Time')  # Title of the graph

# Display the plot
plt.show()
```
