```python
# Example: Plotting a histogram
df['baseFee'].hist()

plt.xlabel('Base Fee')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis
plt.title('Base Fee Histogram In A Day')  # Title of the graph

plt.show()

# Example: Creating a scatter plot
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.plot(df.index, df['baseFee'])  # Replace 'baseFee' with the exact column name

plt.xlabel('Hour')  # Label for the x-axis
plt.ylabel('Base Fee')  # Label for the y-axis
plt.title('Base Fee In A Day')  # Title of the graph

plt.show()
```
